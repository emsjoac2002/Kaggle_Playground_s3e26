# Colinearity and feature selection

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 

cont_df = df[continuous_variables]

vif_data = pd.DataFrame() 
vif_data["feature"] = continuous_variables 
  
# calculating VIF for each feature 
vif_data["VIF"] = [variance_inflation_factor(cont_df.values, i) 
                          for i in range(len(cont_df.columns))] 

vif_data.set_index("feature", inplace = True)

vif_data["VIF"] = vif_data["VIF"].apply(lambda x: format(x, '.4f'))
vif_data
```

```python
def forward_feature_selection(X, y, model):
    selected_features = []
    best_score = 0

    while True:
        remaining_features = list(set(X.columns) - set(selected_features))

        if not remaining_features:
            break  # No remaining features to add

        scores = []

        for feature in remaining_features:
            candidate_features = selected_features + [feature]
            X_subset = X[candidate_features]

            X_train, X_test, y_train, y_test = train_test_split(X_subset, y, test_size=0.2, random_state=42)

            model.fit(X_train, y_train)
            y_pred = model.predict_proba(X_test)

            score = log_loss(y_test, y_pred)

            scores.append((feature, score))

        # Select the feature with the best score
        best_feature, best_score = min(scores, key=lambda x: x[1])

        # Add the best feature to the selected features
        selected_features.append(best_feature)
        
        print(f"Added feature: {best_feature}, Score: {best_score}")

    return selected_features
```

# Column transformation

```python
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('train.csv')
df.drop("id",axis = 1, inplace = True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['Edema'], prefix='Edema')

# Define columns with binary variables
binary_columns = ['Ascites', 'Hepatomegaly','Spiders']

# Map 'Y' to 1 and 'N' to 0 for the specified columns
df[binary_columns] = df[binary_columns].replace({'Y': 1, 'N': 0})
df['Sex'] = df['Sex'].replace({'M': 1, 'F': 0})
df["Stage"] = df["Stage"].astype('int')
df['Drug'] = df['Drug'].replace({'D-penicillamine': 1, 'Placebo': 0})
df['Status'] = df['Status'].replace({'C': 0, 'CL': 1, 'D': 2})
```

```python
df.head()
```

```python
power_list = ["Copper","Alk_Phos","Cholesterol","Tryglicerides","Bilirubin","Prothrombin","SGOT"]
standard_list = ["Albumin","Platelets"]
minmax_list = ["Age","N_Days"]

total_transf_cols = power_list + standard_list + minmax_list

preproc_pipeline = ColumnTransformer([
    ("skewed_transform",PowerTransformer(),power_list),
    ("normal_transform",StandardScaler(),standard_list),
    ("minmax_transform",MinMaxScaler(),minmax_list)
])
```

```python
df_post_proc = df.copy()
df_post_proc[total_transf_cols] = preproc_pipeline.fit_transform(df_post_proc[total_transf_cols])
```

```python
def transform_df(df,pipeline,pipeline_list,test = False):
    
    df = pd.get_dummies(df, columns=['Edema'], prefix='Edema')

    binary_columns = ['Ascites', 'Hepatomegaly','Spiders']

    df[binary_columns] = df[binary_columns].replace({'Y': 1, 'N': 0})
    df['Sex'] = df['Sex'].replace({'M': 1, 'F': 0})
    df["Stage"] = df["Stage"].astype('int')
    df['Drug'] = df['Drug'].replace({'D-penicillamine': 1, 'Placebo': 0})
    
    if test:
        df[pipeline_list] = pipeline.transform(df[pipeline_list])
    else:
        df[pipeline_list] = pipeline.fit_transform(df[pipeline_list])
        df = pd.get_dummies(df, columns=['Status'], prefix='Status')
        
    return df, pipeline
```

# Modelling

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('train.csv')
df.drop("id",axis = 1, inplace = True)
```

```python
power_list = ["Copper","Alk_Phos","Cholesterol","Tryglicerides","Bilirubin","Prothrombin","SGOT"]
standard_list = ["Albumin","Platelets"]
minmax_list = ["Age","N_Days"]

total_transf_cols = power_list + standard_list + minmax_list

preproc_pipeline = ColumnTransformer([
    ("skewed_transform",PowerTransformer(),power_list),
    ("normal_transform",StandardScaler(),standard_list),
    ("minmax_transform",MinMaxScaler(),minmax_list)
])
```

```python
X, pipeline = transform_df(df,preproc_pipeline,total_transf_cols)

status_output = [col for col in X.columns if "Status" in col]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.drop(status_output,axis = 1), X[status_output],
                                                    test_size=0.2, random_state=42)
```

```python
# Define the XGBoost classifier
xgb_model = XGBClassifier()

# Define the parameter grid for grid search
param_grid = {
    'learning_rate': [0.1],
    'n_estimators': [200],
    'max_depth': [3],
    'gamma': [0],
    'subsample': [0.8]
}

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, scoring='neg_log_loss', cv=5, n_jobs=-1)

# Fit the model to the training data
grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters: ", grid_search.best_params_)
print("Best log-loss: ", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
```
